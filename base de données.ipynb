# [dbo].[GetAccessCompetencyList]

```sql
/****** Object:  StoredProcedure [dbo].[GetAccessCompetencyList]    Script Date: 18/08/2025 16:20:45 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- 1) Liste des AccessCompetency pour un utilisateur donné
CREATE PROCEDURE [dbo].[GetAccessCompetencyList]
    @UserID BIGINT
AS
BEGIN
    SET NOCOUNT ON;

    SELECT
        AC.ID                   AS AccessCompetencyID,
        AC.Name                 AS CompetencyName,
       
        ISNULL(UAC.userId, 0)   AS IsActive,
        UAC.LastEditBy          AS LastEditedBy,
        UAC.LastEditAt          AS LastEditedAt
    FROM dbo.ACCESS_COMPETENCY AS AC
    LEFT JOIN dbo.USER_ACCESS_COMPETENCY AS UAC
        ON AC.ID   = UAC.AccessCompetencyID
       AND UAC.UserID = @UserID
    WHERE AC.IsActive = 1
    ORDER BY AC.Name;
END
GO
```

# [dbo].[sp_UpdateUser]

```sql
/****** Object:  StoredProcedure [dbo].[sp_UpdateUser]    Script Date: 18/08/2025 16:20:45 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- 2) Update user (keep password if NULL/empty)
CREATE   PROCEDURE [dbo].[sp_UpdateUser]
  @Id           INT,
  @IsActive     BIT,
  @UserName     NVARCHAR(100),
  @FirstName    NVARCHAR(100),
  @LastName     NVARCHAR(100),
  @Email        NVARCHAR(255),
  @Phone        NVARCHAR(50) = NULL,
  @TeamId       INT = NULL,
  @PasswordHash NVARCHAR(200) = NULL,
  @By           NVARCHAR(100) = NULL
AS
BEGIN
  SET NOCOUNT ON;
  IF @By IS NULL SET @By = SUSER_SNAME();

  UPDATE u
  SET
    u.IsActive     = @IsActive,
    u.UserName     = @UserName,
    u.FirstName    = @FirstName,
    u.LastName     = @LastName,
    u.Email        = @Email,
    u.Phone        = @Phone,
    u.TeamId       = @TeamId,
    u.PasswordHash = CASE WHEN NULLIF(LTRIM(RTRIM(@PasswordHash)), N'') IS NULL THEN u.PasswordHash ELSE @PasswordHash END,
    u.LastEditAt   = GETDATE(),
    u.LastEditBy   = @By
  FROM dbo.[User] u
  WHERE u.Id = @Id;
END

GO
```
